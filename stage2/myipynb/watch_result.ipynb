```python
import _pickle as pickle
import re
import pandas as pd
import json
```

```python
qulac = json.load(open('../data/qulac.json', 'r', encoding='utf-8'))
qulac = pd.DataFrame(qulac)
```

```python
result = pd.read_csv('../data/dev_small.len1.result', sep=' ', header=None).fillna('no_q')
cid2question = {}
for cid, q in result[[0, 2]].values:
    cid2question[cid] = q
```

```python
devset[98392]
```

```python
noq_cnt = 0
for key, val in devset.items():
    val['predict_question'] = cid2question[val['context_id']]
    if val['predict_question'] == 'no_q' or len(val['predict_question']) == 0:
        noq_cnt += 1
    
    t1 = val['topic_id'] == qulac['topic_id']
    t2 = val['question'] == qulac['question']
    t3 = val['answer'] == qulac['answer']
    tmp_df = qulac[t1 & t2]
    
    if len(tmp_df) > 0:
        val['topic_desc'] = tmp_df['topic_desc'].values[0]
        val['facet_desc'] = tmp_df['facet_desc'].values[0]
    else:
        val['topic_desc'] = '-'
        val['facet_desc'] = '-'
    
    print(json.dumps(val, ensure_ascii=False))
    print('#'*20)
```
